```python
class User:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def say_name(self):
        print(f'Меня зовут, {self.name}')

    def say_age(self):
        print(f'Мне, {self.age}')

    def set_age(self, new_age):
        self.age = new_age
        print(f'Возраст изменен на : {self.age}')


user_1 = User('Alex', 22)
user_2 = User('Mark', 22)

user_1.say_name()
user_1.say_age()
user_1.set_age(23)
user_2.say_name()
user_2.say_age()
user_2.set_age(28)
user_2.say_age()

```
