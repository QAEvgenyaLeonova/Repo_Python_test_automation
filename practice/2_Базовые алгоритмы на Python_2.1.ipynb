```python
# Получить от пользователя оценку(Моя версия)

rate = int(input('Оцените работу оператора от 1 до 5'))

#Проверить что оцунка от 1 до 5

if (rate < 1):
    rate = 1
if (rate > 5):
    rate = 5

#в зависимости от оценки предложить обратную связь
feedback = ''

if rate == 1:
    feedback = input('Расскажите что нам улучшить?')
elif (rate == 2):
    feedback = input('Что вас смутило?')
elif (rate == 3):
    feedback = input('Что улучшить?')
else:
    feedback = ('Спасибо , что вы оставили хорошую оценку')

print(feedback)

```

```python
# Получить от пользователя оценку(Версия учителя)

rate_as_str = input('Оцените работу оператора от 1 до 5')
rate = int(rate_as_str)

#Проверить что оцунка от 1 до 5

if (rate < 1):
    rate = 1

if (rate > 5):
    rate = 5

#в зависимости от оценки предложить обратную связь

feedback = ''

if rate == 1:
    feedback = input('Расскажите что нам улучшить?')
else:
    if rate == 2:
        feedback = input('Что вас смутило?')
    else:
        if rate == 3:
            feedback = input('Что улучшить?')


print(feedback)
```
