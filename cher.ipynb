```python
def eat_snacks(*args):
    print("Ура! В корзине:")
    for snack in args:
        print(f" - {snack}")

# Вызываем функцию с разным количеством аргументов
eat_snacks("конфета", "печенье", "торт")
eat_snacks("яблоко", "банан")
eat_snacks("чипсы")
```
